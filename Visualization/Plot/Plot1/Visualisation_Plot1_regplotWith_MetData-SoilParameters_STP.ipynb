##### Plot - regplot - 
STP_TSoil5_1_1_1 = Soil temperature @ 5cm 
STP_TSoil10_1_1_1 = Soil temperature @ 10cm 
STP_TSoil20_1_1_1 = Soil temperature @ 20cm 
STP_TSoil50_1_1_1 = Soil temperature @ 50cm

```python
#plot the different data frames

fig, (axs) = plt.subplots(ncols=2, nrows=2, sharey=True, figsize=(15,8))
fig.suptitle("Soil (STP) Parameters v/s N2O_dry")

sns.regplot(x="STP_TSoil5_1_1_1", y="Flux", data=data_plot1_metSTP, ax=axs[0,0], label="STP_TSoil5_1_1_1", color='#40B790')
sns.regplot(x="STP_TSoil10_1_1_1", y="Flux", data=data_plot1_metSTP, ax=axs[0,1], label="STP_TSoil10_1_1_1", color="#6DA1E5")

sns.regplot(x="STP_TSoil20_1_1_1", y="Flux", data=data_plot1_metSTP, ax=axs[1,0], label="STP_TSoil20_1_1_1", color="#40B790")
sns.regplot(x="STP_TSoil50_1_1_1", y="Flux", data=data_plot1_metSTP, ax=axs[1,1], label="STP_TSoil50_1_1_1", color="#6DA1E5")

axs[0,0].legend()
axs[0,1].legend()
axs[1,0].legend()
axs[1,1].legend()
```

```python
#plot the different data frames

fig, (axs) = plt.subplots(ncols=2, nrows=2, sharey=True, figsize=(15,8))
fig.suptitle("Soil STP Parameters v/s NO2_dry")

sns.regplot(x="STP_TSoil5_1_1_1", y="Flux", data=data_plot1_metSTP, ax=axs[0,0], label="STP_TSoil5_1_1_1", color='#40B790')
sns.regplot(x="STP_TSoil10_1_1_1", y="Flux", data=data_plot1_metSTP, ax=axs[0,1], label="STP_TSoil10_1_1_1", color="#6DA1E5")

sns.regplot(x="STP_TSoil20_1_1_1", y="Flux", data=data_plot1_metSTP, ax=axs[1,0], label="STP_TSoil20_1_1_1", color="#40B790")
sns.regplot(x="STP_TSoil50_1_1_1", y="Flux", data=data_plot1_metSTP, ax=axs[1,1], label="STP_TSoil50_1_1_1", color="#6DA1E5")

fig.text(0.73, 1.0, ' STP_TSoil_2_1_1_1 = Soil temperature @ 2cm \n \
STP_TSoil5_1_1_1 = Soil temperature @ 5cm \n \
STP_TSoil10_1_1_1 = Soil temperature @ 10cm \n \
STP_TSoil20_1_1_1 = Soil temperature @ 20cm \n \
STP_TSoil50_1_1_1 = Soil temperature @ 50cm', \
fontsize=16, color='darkblue'
)
axs[0,0].legend()
axs[0,1].legend()
axs[1,0].legend()
axs[1,1].legend()

#Save boxplot as a file
#output file name
output_file_name = "Plot1/Plot1_gasflux_N2Odry_SoilSTPMetData_regplot.png" 
plot_file_name = plot_file_path + output_file_name
# save as png
plt.savefig(plot_file_name, dpi=50, bbox_inches='tight')
```
