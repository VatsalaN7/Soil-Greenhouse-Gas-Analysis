### Separate the gases

#### N2O_dry

```python
data_plot1_metA_N2Odry = data_plot1_metA[data_plot1_metA['ghg'] == 'N2O_dry']
```

```python
#data_plot1_metA_N2Odry
```

```python
#plot the different data frames

fig, (axs) = plt.subplots(ncols=2, nrows=3, sharey=True, figsize=(15,8))
fig.suptitle("Air Parameters v/s N2O_dry", fontsize=22)

sns.regplot(x="G_1_1_1", y="Flux", data=data_plot1_metA_N2Odry, ax=axs[0,0], label="G_1_1_1", color='#40B790')
sns.regplot(x="G_2_1_1", y="Flux", data=data_plot1_metA_N2Odry, ax=axs[0,1], label="G_2_1_1", color="#40B790")

sns.regplot(x="Ta_1_1_1", y="Flux", data=data_plot1_metA_N2Odry, ax=axs[1,0], label="Ta_1_1_1", color="#6DA1E5")
sns.regplot(x="RH_1_1_1", y="Flux", data=data_plot1_metA_N2Odry, ax=axs[1,1], label="RH_1_1_1", color="#6DA1E5")

sns.regplot(x="Pa_1_1_1", y="Flux", data=data_plot1_metA_N2Odry, ax=axs[2,0], label="Pa_1_1_1", color="#9E6EB3")
sns.regplot(x="P_1_1_1", y="Flux", data=data_plot1_metA_N2Odry, ax=axs[2,1], label="P_1_1_1", color="#40B790")

fig.text(0.83, 0.98,' G_1_1_1 = Soil heat flux \n \
G_2_1_1 = Soil heat flux \n \
Ta_1_1_1 = Air temperature \n \
RH_1_1_1 = Humidity \n \
Pa_1_1_1 = Atmospheric Pressure \n \
P_1_1_1 = Precipitation', fontsize=16, color='darkblue'
)  

axs[0,0].legend()
axs[0,1].legend()
axs[1,0].legend()
axs[1,1].legend()
axs[2,0].legend()
axs[2,1].legend()

#Save boxplot as a file
#output file name
output_file_name = "Plot1/Plot1_gasflux_N2Odry_AirMetData_regplot.png" 
plot_file_name = plot_file_path + output_file_name
# save as png
plt.savefig(plot_file_name, dpi=50, bbox_inches='tight')
```

### Note: N2O_dry - does not seem to have a linear relation with the parameters
