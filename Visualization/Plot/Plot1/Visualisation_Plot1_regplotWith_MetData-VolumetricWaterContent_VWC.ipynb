##### Plot - regplot - 
VWC_1_1_1 = Volumetric water capacity @ 5cm, VWC_2_1_1 = Volumetric water capacity @ 10cm, VWC_3_1_1 = Volumetric water capacity @ 15cm, VWC_4_1_1 = Volumetric water capacity @ 25cm

```python
#plot the different data frames

fig, (axs) = plt.subplots(ncols=2, nrows=2, sharey=True, figsize=(15,8))
fig.suptitle("Volumetric Water Content Parameters v/s Flux")

sns.regplot(x="VWC_1_1_1", y="Flux", data=data_plot1_metVWC, ax=axs[0,0], label="VWC_1_1_1", color='#40B790')
sns.regplot(x="VWC_2_1_1", y="Flux", data=data_plot1_metVWC, ax=axs[0,1], label="VWC_2_1_1", color="#6DA1E5")

sns.regplot(x="VWC_3_1_1", y="Flux", data=data_plot1_metVWC, ax=axs[1,0], label="VWC_3_1_1", color="#40B790")
sns.regplot(x="VWC_4_1_1", y="Flux", data=data_plot1_metVWC, ax=axs[1,1], label="VWC_4_1_1", color="#6DA1E5")

axs[0,0].legend()
axs[0,1].legend()
axs[1,0].legend()
axs[1,1].legend()
```

##### Separate the gases 

##### N2O_dry

```python
data_plot1_metVWC_N2Odry = data_plot1_metVWC[data_plot1_metVWC['ghg'] == 'N2O_dry']
```

```python
data_plot1_metVWC_N2Odry
```

```python
#plot the different data frames

fig, (axs) = plt.subplots(ncols=2, nrows=2, sharey=True, figsize=(15,8))
fig.suptitle("Volumetric Water Content Parameters v/s N2O_dry")

sns.regplot(x="VWC_1_1_1", y="Flux", data=data_plot1_metVWC_N2Odry, ax=axs[0,0], label="VWC_1_1_1", color='#40B790')
sns.regplot(x="VWC_2_1_1", y="Flux", data=data_plot1_metVWC_N2Odry, ax=axs[0,1], label="VWC_2_1_1", color="#6DA1E5")

sns.regplot(x="VWC_3_1_1", y="Flux", data=data_plot1_metVWC_N2Odry, ax=axs[1,0], label="VWC_3_1_1", color="#40B790")
sns.regplot(x="VWC_4_1_1", y="Flux", data=data_plot1_metVWC_N2Odry, ax=axs[1,1], label="VWC_4_1_1", color="#6DA1E5")

fig.text(0.73, 1.0, ' VWC_1_1_1 = Volumetric water capacity @ 5cm \n \
VWC_2_1_1 = Volumetric water capacity @ 10cm \n \
VWC_3_1_1 = Volumetric water capacity @ 15cm \n \
VWC_4_1_1 = Volumetric water capacity @ 25cm', \
fontsize=16, color='darkblue'
)

axs[0,0].legend()
axs[0,1].legend()
axs[1,0].legend()
axs[1,1].legend()

#Save boxplot as a file
#output file name
output_file_name = "Plot1/Plot1_gasflux_N2Odry_SoilVWCMetData_regplot.png" 
plot_file_name = plot_file_path + output_file_name
# save as png
plt.savefig(plot_file_name, dpi=50, bbox_inches='tight')
```

##### Note: There does not seem to be any linear relation between N2O_dry flux and Volumetric Water Content
