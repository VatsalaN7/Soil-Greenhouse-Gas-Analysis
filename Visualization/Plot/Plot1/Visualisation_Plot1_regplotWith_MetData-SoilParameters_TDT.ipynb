##### Plot - regplot - 
Soil temperature @ 5cm, Soil temperature @ 10cm, Soil temperature @ 15cm, Soil temperature @ 25cm

```python
#plot the different data frames

fig, (axs) = plt.subplots(ncols=2, nrows=2, sharey=True, figsize=(15,8))
fig.suptitle("Soil (TDT) Parameters v/s Flux")

sns.regplot(x="TDT1_Ts_1_1_1", y="Flux", data=data_plot1_metTDT, ax=axs[0,0], label="TDT1_Ts_1_1_1", color='#40B790')
sns.regplot(x="TDT2_Ts_1_1_1", y="Flux", data=data_plot1_metTDT, ax=axs[0,1], label="TDT2_Ts_1_1_1", color="#6DA1E5")

sns.regplot(x="TDT3_Ts_1_1_1", y="Flux", data=data_plot1_metTDT, ax=axs[1,0], label="TDT3_Ts_1_1_1", color="#40B790")
sns.regplot(x="TDT4_Ts_1_1_1", y="Flux", data=data_plot1_metTDT, ax=axs[1,1], label="TDT4_Ts_1_1_1", color="#6DA1E5")

axs[0,0].legend()
axs[0,1].legend()
axs[1,0].legend()
axs[1,1].legend()
```

##### Separate the gases 

##### N2O_dry

```python
data_plot1_metTDT_N2Odry = data_plot1_metTDT[data_plot1_metTDT['ghg'] == 'N2O_dry']
```

```python
data_plot1_metTDT_N2Odry
```

```python
#plot the different data frames

fig, (axs) = plt.subplots(ncols=2, nrows=2, sharey=True, figsize=(15,8))
fig.suptitle("Soil (TDT) Parameters v/s N2O_dry")

sns.regplot(x="TDT1_Ts_1_1_1", y="Flux", data=data_plot1_metTDT_N2Odry, ax=axs[0,0], label="TDT1_Ts_1_1_1", color='#40B790')
sns.regplot(x="TDT2_Ts_1_1_1", y="Flux", data=data_plot1_metTDT_N2Odry, ax=axs[0,1], label="TDT2_Ts_1_1_1", color="#6DA1E5")

sns.regplot(x="TDT3_Ts_1_1_1", y="Flux", data=data_plot1_metTDT_N2Odry, ax=axs[1,0], label="TDT3_Ts_1_1_1", color="#40B790")
sns.regplot(x="TDT4_Ts_1_1_1", y="Flux", data=data_plot1_metTDT_N2Odry, ax=axs[1,1], label="TDT4_Ts_1_1_1", color="#6DA1E5")

fig.text(0.73, 1.0, ' TDT1_Ts_1_1_1 = Soil temperature @ 5cm \n \
TDT2_Ts_1_1_1 = Soil temperature @ 10cm \n \
TDT3_Ts_1_1_1 = Soil temperature @ 15cm \n \
TDT4_Ts_1_1_1 = Soil temperature @ 25cm', \
fontsize=16, color='darkblue'
)

axs[0,0].legend()
axs[0,1].legend()
axs[1,0].legend()
axs[1,1].legend()

#Save boxplot as a file
#output file name
output_file_name = "Plot1/Plot1_gasflux_N2Odry_SoilTDTMetData_regplot.png" 
plot_file_name = plot_file_path + output_file_name
# save as png
plt.savefig(plot_file_name, dpi=50, bbox_inches='tight')
```

##### Note: There does not seem to be any linear relation between N2O_dry flux and Soil TDT parameters
