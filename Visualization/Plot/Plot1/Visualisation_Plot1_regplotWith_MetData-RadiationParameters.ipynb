#### Functions to draw linear regression models - regplot, lmplot

##### PPFD_1_1_1 = Photosynthetic density, SWin_1_1_1 = Shortwave incoming radiation, SWout_1_1_1 = Shortwave outgoing radiation, LWin_1_1_1 = Longwave incoming radiation, LWout_1_1_1 = Longwave outgoing radiation, Rnet_1_1_1 = Net radiation

```python
sns.regplot(x="PPFD_1_1_1", y="Flux", data=data_plot1_metR)
```

##### SWin_1_1_1 = Shortwave incoming radiation

```python
sns.regplot(x="SWin_1_1_1", y="Flux", data=data_plot1_metR)
```

```python
#plot the different data frames

fig, (ax1, ax2) = plt.subplots(ncols=1, nrows=2, sharey=True, figsize=(10,4))

sns.regplot(x="PPFD_1_1_1", y="Flux", data=data_plot1_metR, ax=ax1, label="PPFD_1_1_1", color='#40B790')
sns.regplot(x="SWin_1_1_1", y="Flux", data=data_plot1_metR, ax=ax2, label="SWin_1_1_1", color="#6DA1E5")
ax1.legend()
ax2.legend()
```

```python
#plot the different data frames

fig, (axs) = plt.subplots(ncols=2, nrows=3, sharey=True, figsize=(15,8))
fig.suptitle("Radiation Parameters v/s Flux")

sns.regplot(x="PPFD_1_1_1", y="Flux", data=data_plot1_metR, ax=axs[0,0], label="PPFD_1_1_1", color='#40B790')
sns.regplot(x="SWin_1_1_1", y="Flux", data=data_plot1_metR, ax=axs[0,1], label="SWin_1_1_1", color="#40B790")

sns.regplot(x="SWout_1_1_1", y="Flux", data=data_plot1_metR, ax=axs[1,0], label="SWout_1_1_1", color="#40B790")
sns.regplot(x="LWin_1_1_1", y="Flux", data=data_plot1_metR, ax=axs[1,1], label="LWin_1_1_1", color="#6DA1E5")

sns.regplot(x="LWout_1_1_1", y="Flux", data=data_plot1_metR, ax=axs[2,0], label="LWout_1_1_1", color="#9E6EB3")
sns.regplot(x="Rnet_1_1_1", y="Flux", data=data_plot1_metR, ax=axs[2,1], label="Rnet_1_1_1", color="#40B790")


axs[0,0].legend()
axs[0,1].legend()
axs[1,0].legend()
axs[1,1].legend()
axs[2,0].legend()
axs[2,1].legend()
```

##### Separate the gases

##### N2O_dry

```python
data_plot1_metR_N2Odry = data_plot1_metR[data_plot1_metR['ghg'] == 'N2O_dry']
```

```python
data_plot1_metR_N2Odry
```

```python
#plot the different data frames

fig, (axs) = plt.subplots(ncols=2, nrows=3, sharey=False, sharex=False, figsize=(15,8))
fig.suptitle("Radiation Parameters v/s N2O_dry")

sns.regplot(x="PPFD_1_1_1", y="Flux", data=data_plot1_metR_N2Odry, ax=axs[0,0], label="PPFD_1_1_1", color='#40B790')
sns.regplot(x="SWin_1_1_1", y="Flux", data=data_plot1_metR_N2Odry, ax=axs[0,1], label="SWin_1_1_1", color="#40B790")

sns.regplot(x="SWout_1_1_1", y="Flux", data=data_plot1_metR_N2Odry, ax=axs[1,0], label="SWout_1_1_1", color="#40B790")
sns.regplot(x="LWin_1_1_1", y="Flux", data=data_plot1_metR_N2Odry, ax=axs[1,1], label="LWin_1_1_1", color="#6DA1E5")

sns.regplot(x="LWout_1_1_1", y="Flux", data=data_plot1_metR_N2Odry, ax=axs[2,0], label="LWout_1_1_1", color="#9E6EB3")
sns.regplot(x="Rnet_1_1_1", y="Flux", data=data_plot1_metR_N2Odry, ax=axs[2,1], label="Rnet_1_1_1", color="#40B790")

fig.text(0.73, 0.98,' PPFD_1_1_1 = Photosynthetic photon flux density \n \
SWin_1_1_1 = Shortwave incoming radiation \n \
SWout_1_1_1 = Shortwave outgoing radiation \n \
LWin_1_1_1 = Longwave incoming radiation \n \
LWout_1_1_1 = Longwave outgoing radiation \n \
Rnet_1_1_1 = Net radiation', fontsize=16, color='darkblue'
)   

axs[0,0].legend()
axs[0,1].legend()
axs[1,0].legend()
axs[1,1].legend()
axs[2,0].legend()
axs[2,1].legend()

#Save boxplot as a file
#output file name
output_file_name = "Plot1/Plot1_gasflux_N2Odry_RadiationMetData_regplot.png" 
plot_file_name = plot_file_path + output_file_name
# save as png
plt.savefig(plot_file_name, dpi=50, bbox_inches='tight')
```

##### Note: N2O_dry - does not seem to have a linear relation with Radiation parameters
